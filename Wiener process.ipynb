```python
import scipy

N = 100
T = 1.0
dt = T/N

B = scipy.zeros(N+1)
t = scipy.linspace (0 , T , N +1)
B[1: N +1] = scipy.cumsum(scipy.sqrt(dt)*scipy.random.standard_normal(N))


import matplotlib.pyplot as plt

plt.figure(num=None, figsize=(10, 6), dpi=200, facecolor='w', edgecolor='k')

for i in range(1,100):
    B[1: N +1] = scipy.cumsum(scipy.sqrt(dt)*scipy.random.standard_normal(N))
    plt.plot(t, B)

plt.show()
```
